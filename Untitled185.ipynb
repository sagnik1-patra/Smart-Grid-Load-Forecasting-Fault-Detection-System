```python
import os, json, pickle, yaml, numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import load_model

# ------------------------------------------------------------
# Paths
# ------------------------------------------------------------
BASE = r"C:\Users\NXTWAVE\Downloads\Smart Grid Load Forecasting & Fault Detection System"
DATA = os.path.join(BASE, "archive", "Data_for_UCI_named.csv")
MODEL = os.path.join(BASE, "ais_csa_energi_model.h5")
SCALER = os.path.join(BASE, "ais_csa_energi_scaler.pkl")
CFG = os.path.join(BASE, "ais_csa_energi_config.yaml")
CSV_OUT = os.path.join(BASE, "hybrid_ais_csa_results.csv")
JSON_OUT = os.path.join(BASE, "hybrid_ais_csa_prediction.json")

# ------------------------------------------------------------
# Safe-load or rebuild YAML
# ------------------------------------------------------------
def load_or_repair_yaml(path, model):
    try:
        with open(path, "r") as f:
            text = f.read()
        # Remove python tags but don't try to keep broken blocks
        import re
        text = re.sub(r"!!python[^\n]+", "", text)
        text = re.sub(r":numpy[^\n]+", "", text)
        cfg = yaml.safe_load(text)
        if not isinstance(cfg, dict):
            raise ValueError
        return cfg
    except Exception as e:
        print(f"[WARN] Config unreadable ({e}); rebuilding minimal config...")
        # Build a generic config using model info
        return {
            "Hybrid_Model": "AIS + CSA Optimized LSTM",
            "Hidden_Layers": len(model.layers),
            "Total_Params": int(model.count_params()),
            "Note": "Auto-generated clean config due to YAML corruption"
        }

# ------------------------------------------------------------
# Load model + scalers
# ------------------------------------------------------------
print("[INFO] Loading model and scalers...")
model = load_model(MODEL)
scalers = pickle.load(open(SCALER, "rb"))
scaler_x, scaler_y = scalers["x"], scalers["y"]

config = load_or_repair_yaml(CFG, model)

# rewrite a clean YAML for future runs
with open(CFG, "w") as f:
    yaml.safe_dump(config, f)

print("[INFO] Model & config ready ✅")
print(config)

# ------------------------------------------------------------
# Load dataset
# ------------------------------------------------------------
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA)
target_col = "stab"
if target_col not in df.columns:
    target_col = [c for c in df.columns if "stab" in c.lower()][0]
if "stabf" in df.columns:
    df.drop("stabf", axis=1, inplace=True)

X = df.drop(target_col, axis=1).values
y = df[target_col].values.reshape(-1, 1)

# ------------------------------------------------------------
# Predict
# ------------------------------------------------------------
X_scaled = scaler_x.transform(X).reshape((X.shape[0], 1, X.shape[1]))
print("[INFO] Generating predictions...")
y_pred_scaled = model.predict(X_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

# ------------------------------------------------------------
# Metrics
# ------------------------------------------------------------
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
print("\n========== RESULTS ==========")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R²   : {r2:.4f}")
print("=============================")

# ------------------------------------------------------------
# Fault probability heuristic
# ------------------------------------------------------------
deviation = np.abs(y - y_pred)
fault_prob = np.mean(deviation > 1.5 * rmse)
if fault_prob < 0.05:
    status = "Grid stable — within safe limits"
elif fault_prob < 0.15:
    status = "Moderate fluctuations — monitor"
else:
    status = "⚠️  High fault probability — inspect grid"

# ------------------------------------------------------------
# Save outputs
# ------------------------------------------------------------
pd.DataFrame({
    "Actual_Load": y.flatten(),
    "Predicted_Load": y_pred.flatten(),
    "Absolute_Error": np.abs(y.flatten() - y_pred.flatten())
}).to_csv(CSV_OUT, index=False)

summary = {
    "Hybrid_Model": config.get("Hybrid_Model", "AIS + CSA Optimized LSTM"),
    "Metrics": {"MAE": float(mae), "RMSE": float(rmse), "R2": float(r2)},
    "Fault_Probability": float(fault_prob),
    "Recommendations": [
        status,
        f"Avg deviation ≈ {np.mean(deviation):.2f} units",
        "Recalibrate if RMSE > 5.0"
    ]
}
with open(JSON_OUT, "w") as f:
    json.dump(summary, f, indent=4)

print(f"[✅] Results saved to: {BASE}")

# ------------------------------------------------------------
# Visualization
# ------------------------------------------------------------
plt.figure(figsize=(10,5))
plt.plot(y[:200], label="Actual", linewidth=2)
plt.plot(y_pred[:200], label="Predicted", linewidth=2)
plt.title("Actual vs Predicted Load (Sample 200)")
plt.xlabel("Samples"); plt.ylabel("Load (kW)")
plt.legend(); plt.grid(True)
plt.show()

print(json.dumps(summary, indent=4))
```
