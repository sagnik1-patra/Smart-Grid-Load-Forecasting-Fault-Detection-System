```python
import os
import json
import yaml
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
import pickle

# ============================================================
# 1ï¸â£ Paths
# ============================================================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Grid Load Forecasting & Fault Detection System"
DATA_PATH = os.path.join(BASE_DIR, "archive", "Data_for_UCI_named.csv")

MODEL_PATH = os.path.join(BASE_DIR, "hybrid_energi_model.h5")
SCALER_PATH = os.path.join(BASE_DIR, "hybrid_energi_scaler.pkl")
CONFIG_PATH = os.path.join(BASE_DIR, "hybrid_energi_config.yaml")
REPORT_PATH = os.path.join(BASE_DIR, "hybrid_energi_report.json")
RESULTS_PATH = os.path.join(BASE_DIR, "hybrid_energi_results.csv")

ACC_GRAPH = os.path.join(BASE_DIR, "hybrid_energi_accuracy_graph.png")
HEATMAP = os.path.join(BASE_DIR, "hybrid_energi_heatmap.png")
PRED_GRAPH = os.path.join(BASE_DIR, "hybrid_energi_prediction_graph.png")
RESULT_GRAPH = os.path.join(BASE_DIR, "hybrid_energi_result_graph.png")

# ============================================================
# 2ï¸â£ Load & Inspect
# ============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)
print(f"[INFO] Shape: {df.shape}")
print(f"[INFO] Columns: {list(df.columns)}")

target_col = "stab"
if target_col not in df.columns:
    target_col = [c for c in df.columns if 'stab' in c.lower()][0]

if 'stabf' in df.columns:
    df = df.drop('stabf', axis=1)

# ============================================================
# 3ï¸â£ Preprocess
# ============================================================
X = df.drop(target_col, axis=1).values
y = df[target_col].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_train_scaled = scaler_x.fit_transform(X_train)
X_test_scaled = scaler_x.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)

X_train_scaled = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_scaled = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# ============================================================
# 4ï¸â£ Hybrid AIS + PSO Optimizer
# ============================================================

def fitness(hidden, dropout):
    model = Sequential([
        LSTM(int(hidden), input_shape=(1, X_train.shape[1]), return_sequences=False),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train_scaled, y_train_scaled, epochs=10, batch_size=32, verbose=0)
    y_pred = model.predict(X_test_scaled)
    return np.sqrt(mean_squared_error(y_test_scaled, y_pred))

def hybrid_ais_pso(n_particles=6, n_iter=5):
    # Initialize population (particles)
    particles = [{
        "hidden": random.randint(32, 128),
        "dropout": random.uniform(0.1, 0.5),
        "velocity_h": 0,
        "velocity_d": 0
    } for _ in range(n_particles)]

    best_global = None
    best_rmse = float('inf')

    for iteration in range(n_iter):
        print(f"[HYBRID AIS+PSO] Iteration {iteration+1}/{n_iter}")

        for p in particles:
            rmse = fitness(p["hidden"], p["dropout"])
            if rmse < best_rmse:
                best_rmse = rmse
                best_global = p.copy()

        # PSO Update
        for p in particles:
            w = 0.7
            c1, c2 = 1.4, 1.4
            p["velocity_h"] = (w * p["velocity_h"]
                               + c1 * random.random() * (best_global["hidden"] - p["hidden"])
                               + c2 * random.random() * (best_global["hidden"] - p["hidden"]))
            p["velocity_d"] = (w * p["velocity_d"]
                               + c1 * random.random() * (best_global["dropout"] - p["dropout"])
                               + c2 * random.random() * (best_global["dropout"] - p["dropout"]))

            p["hidden"] += p["velocity_h"]
            p["dropout"] += p["velocity_d"]

            p["hidden"] = int(np.clip(p["hidden"], 16, 256))
            p["dropout"] = np.clip(p["dropout"], 0.05, 0.5)

        # AIS (immune cloning + mutation)
        clones = []
        for p in particles:
            clone = p.copy()
            clone["hidden"] = int(np.clip(clone["hidden"] + random.uniform(-8, 8), 16, 256))
            clone["dropout"] = np.clip(clone["dropout"] + random.uniform(-0.05, 0.05), 0.05, 0.5)
            clones.append(clone)

        # Evaluate clones
        for c in clones:
            rmse = fitness(c["hidden"], c["dropout"])
            if rmse < best_rmse:
                best_rmse = rmse
                best_global = c.copy()

    return best_global, best_rmse

print("[INFO] Running Hybrid AIS + PSO optimization...")
best_params, best_rmse = hybrid_ais_pso(n_particles=5, n_iter=3)
print("[INFO] Best Parameters:", best_params)

# ============================================================
# 5ï¸â£ Final Model Training
# ============================================================
hidden = best_params["hidden"]
dropout = best_params["dropout"]

model = Sequential([
    LSTM(hidden, input_shape=(1, X_train.shape[1]), return_sequences=False),
    Dropout(dropout),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
es = EarlyStopping(patience=5, restore_best_weights=True)

history = model.fit(
    X_train_scaled, y_train_scaled,
    validation_data=(X_test_scaled, y_test_scaled),
    epochs=50, batch_size=32, verbose=1, callbacks=[es]
)

# ============================================================
# 6ï¸â£ Evaluation
# ============================================================
y_pred_scaled = model.predict(X_test_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("\n=================== RESULTS ===================")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"RÂ²   : {r2:.4f}")
print("===============================================")

# ============================================================
# 7ï¸â£ Save Artifacts
# ============================================================
pickle.dump({"x": scaler_x, "y": scaler_y}, open(SCALER_PATH, "wb"))
model.save(MODEL_PATH)

config = {
    "Hybrid_Model": "AIS + PSO Optimized LSTM",
    "Hidden": hidden,
    "Dropout": dropout,
    "Optimizer": "adam",
    "Epochs": len(history.history["loss"])
}
with open(CONFIG_PATH, "w") as f:
    yaml.dump(config, f)

report = {"MAE": float(mae), "RMSE": float(rmse), "R2": float(r2), "Best_Params": best_params}
with open(REPORT_PATH, "w") as f:
    json.dump(report, f, indent=4)

results_df = pd.DataFrame({"Actual": y_test.flatten(), "Predicted": y_pred.flatten()})
results_df.to_csv(RESULTS_PATH, index=False)

# ============================================================
# 8ï¸â£ Visualization & Save
# ============================================================

# Accuracy Graph
plt.figure()
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend(); plt.title("Training vs Validation Loss")
plt.savefig(ACC_GRAPH)

# Hybrid Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=False, cmap="viridis")
plt.title("Hybrid EnergiSense Correlation Heatmap")
plt.savefig(HEATMAP)

# Prediction Graph
plt.figure()
plt.plot(y_test[:200], label='Actual')
plt.plot(y_pred[:200], label='Predicted')
plt.legend(); plt.title("Actual vs Predicted Load")
plt.savefig(PRED_GRAPH)

# Result Graph
plt.figure()
sns.barplot(x=['MAE','RMSE','RÂ²'], y=[mae, rmse, r2])
plt.title("Hybrid AIS+PSO Model Metrics")
plt.savefig(RESULT_GRAPH)

plt.show()
print(f"[â] All hybrid artifacts saved in: {BASE_DIR}")
```
